```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 创建一个示例数据框（可以替换为你自己的数据）
data = {
    'x': [1, 2, 3, 4, 5],
    'y1': [10, 15, 13, 18, 20],
    'y2': [5, 8, 7, 12, 14]
}

# 将数据框转换为Seaborn的数据结构
df = pd.DataFrame(data)

sns.set_theme(style="darkgrid")
# 使用Seaborn的lineplot函数绘制折线图
sns.lineplot(x='x', y='y1', data=df, label='Line 1')
sns.lineplot(x='x', y='y2', data=df, label='Line 2')

# 添加标题和标签
plt.title('Two Lines Plot')
plt.ylabel('FPS')
plt.xlabel('Resolution')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
